```python
from bs4 import BeautifulSoup
import requests
page = 1
words_list = []
while page!= 34:
    url = "https://jisho.org/search/%20%23jlpt-n5%20%23words?page={}".format(page)
    result = requests.get(url)
    doc = BeautifulSoup(result.text, "html.parser")
    jlptn = doc.find_all('span', class_="text")
    for i in jlptn:
        words_list.append(i.text.replace('\n', '').strip())
    page+=1
words_list = [i for i in words_list if i not in ['\n', 'Words']]
print(words_list)
```
