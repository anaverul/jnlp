```python
#regex    
hiragana_full = r'[ぁ-ゟ]'
katakana_full = r'[゠-ヿ]'
kanji = r'[㐀-䶵一-鿋豈-頻]'
radicals = r'[⺀-⿕]'
katakana_half_width = r'[｟-ﾟ]'
alphanum_full = r'[！-～]'
symbols_punct = r'[、-〿]'
misc_symbols = r'[ㇰ-ㇿ㈠-㉃㊀-㋾㌀-㍿]'
ascii_char = r'[ -~]'
```

```python
tokenizer_obj = dictionary.Dictionary().create()
mode = tokenizer.Tokenizer.SplitMode.C
tokens = [m.surface() for m in tokenizer_obj.tokenize(text, mode)]
print(tokens)
```

```python
def split_tokens(tok): 
    split_string = ""
    for i in range(len(tok)):
        if tok[i] in alphanum_full:
            split_string += tokens[i]
        else:
            split_string += " "
            split_string += tokens[i]
    split_string = split_string[1::].replace("\n", "")
    print(split_string)
split_tokens(tokens)
```

```python
def split_words(tok): #splitting after detecting particles that need to be attached
    split_string = ""
    for i in range(len(tok)):
        if tok[i] in alphanum_full or tok[i] in Particles:
            #print(tok[i])
            split_string += tokens[i]
        else:
            #print(tok[i])
            split_string += " "
            split_string += tokens[i]
    split_string = split_string[1::].replace("\n", "")
    print(split_string)
split_words(tokens)
```
